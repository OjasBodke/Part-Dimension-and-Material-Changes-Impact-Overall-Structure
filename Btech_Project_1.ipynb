```python
# ===============================================================
# 🛩️ Boeing 737 Change Management Chatbot + Dashboard (Ultra-Fast)
# ===============================================================

import streamlit as st
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
import matplotlib.pyplot as plt
import networkx as nx
import time

# ---------------------------------------------------------------
# Streamlit Configuration
# ---------------------------------------------------------------
st.set_page_config(page_title="Boeing 737 Change Management", layout="wide")
st.markdown("<h1 style='text-align:center;'>🛩️ Boeing 737 Change Management Dashboard</h1>", unsafe_allow_html=True)
st.markdown("<p style='text-align:center; font-size:18px;'>Explore, Analyze, and Chat with Boeing 737 Change Data</p>", unsafe_allow_html=True)
st.write("---")

# ---------------------------------------------------------------
# Cached Loaders
# ---------------------------------------------------------------
@st.cache_data
def load_dataset(file):
    """Load the CSV dataset."""
    df = pd.read_csv(file)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    return df

@st.cache_resource
def load_model():
    """Load SentenceTransformer model once."""
    return SentenceTransformer("all-MiniLM-L6-v2")

@st.cache_resource
def create_embeddings_cached(combined_text_list):
    """Cache embeddings for same dataset."""
    model = load_model()
    embeddings = model.encode(combined_text_list, show_progress_bar=False)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(np.array(embeddings))
    return model, index, embeddings

# ---------------------------------------------------------------
# 1️⃣ Upload & Preview Dataset
# ---------------------------------------------------------------
st.header("📂 Upload & Preview Dataset")

uploaded_file = st.file_uploader("Upload Boeing 737 dataset (CSV)", type=["csv"])

if uploaded_file is not None:
    df = load_dataset(uploaded_file)
    st.success("✅ Dataset uploaded successfully!")
else:
    st.warning("Please upload your dataset first to continue.")
    st.stop()

st.markdown("#### 📊 Dataset Preview")
st.dataframe(df.head(), use_container_width=True)
st.write("---")

# ---------------------------------------------------------------
# 2️⃣ Create Vector Embeddings (Fast)
# ---------------------------------------------------------------
st.header("🧠 Create Vector Embeddings")

text_columns = st.multiselect(
    "Select columns for embedding (combine textual info):",
    options=df.columns.tolist(),
    default=["Problem_Description", "Change_Type"]
)

# Combine text from selected columns
df["combined_text"] = df[text_columns].astype(str).apply(lambda x: " ".join(x), axis=1)

# Check cache or create embeddings
if "model" not in st.session_state or "index" not in st.session_state or "embeddings" not in st.session_state:
    progress = st.progress(0)
    status_text = st.empty()

    for i in range(0, 100, 15):
        progress.progress(i)
        status_text.text(f"🚀 Creating vector embeddings... {i}%")
        time.sleep(0.1)

    model, index, embeddings = create_embeddings_cached(df["combined_text"].tolist())
    st.session_state.model = model
    st.session_state.index = index
    st.session_state.embeddings = embeddings
    progress.progress(100)
    status_text.text("✅ Vector embeddings created successfully!")
    time.sleep(0.3)
    status_text.empty()
else:
    model = st.session_state.model
    index = st.session_state.index
    embeddings = st.session_state.embeddings
    st.success("⚡ Using cached embeddings — instant response!")

st.write("---")

# ---------------------------------------------------------------
# 3️⃣ Chatbot
# ---------------------------------------------------------------
st.header("💬 Boeing 737 Change Management Chatbot")

user_query = st.text_input("Ask a question about a part, change, or issue:")

if user_query:
    query_embedding = model.encode([user_query])
    distances, indices = index.search(np.array(query_embedding), k=5)
    results = df.iloc[indices[0]]

    # --- Display Results ---
    st.subheader("🔎 Top Relevant Results")
    st.dataframe(results, use_container_width=True)

    # --- Generate Summary ---
    st.markdown("#### 🗒️ Summary of Results")
    summary_text = ""
    for _, row in results.iterrows():
        summary_text += (
            f"- **Change ID:** {row.get('Change_ID', 'N/A')} | "
            f"**Part:** {row.get('Primary_Area', 'N/A')} | "
            f"**Problem:** {row.get('Problem_Description', 'N/A')} | "
            f"**Change Type:** {row.get('Change_Type', 'N/A')} | "
            f"**Reported by:** {row.get('Source_of_Report', 'N/A')}\n"
        )
    st.markdown(summary_text)
st.write("---")

# ---------------------------------------------------------------
# 4️⃣ Impact Analysis
# ---------------------------------------------------------------
st.header("⚙️ Impact Analysis: Attribute Change")

col1, col2 = st.columns(2)
with col1:
    attr = st.selectbox("Select Attribute to Change:", df.columns)
with col2:
    unique_values = df[attr].dropna().unique().tolist()
    selected_value = st.selectbox(f"Select New Value for {attr}:", unique_values)

impacted = df[df[attr] == selected_value]

st.markdown(f"### 🧾 Impacted Entries When `{attr}` changes to `{selected_value}`")
st.dataframe(impacted.head(20), use_container_width=True)

if not impacted.empty:
    st.markdown("#### 🧠 Text Summary of Impacted Entries")
    for _, row in impacted.iterrows():
        st.write(
            f"- **Change ID:** {row.get('Change_ID', 'N/A')} | "
            f"**Part:** {row.get('Primary_Part', row.get('Part_Number', 'N/A'))} | "
            f"**Problem:** {row.get('Problem_Description', 'N/A')} | "
            f"**Change Type:** {row.get('Change_Type', 'N/A')} | "
            f"**Date:** {row.get('Date_Reported', 'N/A')}"
        )
else:
    st.warning("No impacted data found for the selected attribute/value.")
st.write("---")

# ---------------------------------------------------------------
# 5️⃣ Graph Visualization
# ---------------------------------------------------------------
st.header("📈 Graph of Impacted Attributes")

if not impacted.empty:
    G = nx.Graph()
    for _, row in impacted.iterrows():
        part = str(row.get("Primary_Part", row.get("Part_Number", "Unknown Part")))
        G.add_node(part)
        G.add_edge(selected_value, part)

    plt.figure(figsize=(10, 6))
    nx.draw(
        G,
        with_labels=True,
        node_color="lightblue",
        node_size=1200,
        font_size=9,
        font_weight="bold"
    )
    st.pyplot(plt)
else:
    st.warning("No impacted data found for the selected attribute/value.")
st.write("---")

# ---------------------------------------------------------------
# 6️⃣ Summary Statistics
# ---------------------------------------------------------------
st.header("📊 Summary Statistics")
st.dataframe(df.describe(include='all').transpose(), use_container_width=True)
st.write("---")

# ---------------------------------------------------------------
# Footer
# ---------------------------------------------------------------
st.markdown("<p style='text-align:center; color:gray;'>© 2025 Boeing 737 Change Management | AI-Powered Dashboard</p>", unsafe_allow_html=True)
```
